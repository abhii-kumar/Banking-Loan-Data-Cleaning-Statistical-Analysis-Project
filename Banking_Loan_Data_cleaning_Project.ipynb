```python
numeric_columns=[]
for i in df.columns:
    if df[i].dtype!="object" and i not in ["PROSPECTID","Approved_Flag"]:  
        numeric_columns.append(i)

dfn = df[numeric_columns]
dfn
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
total_columns = dfn.shape[1]
column_to_be_kept = []
column_index = 0

for i in range(0,total_columns):
    vif_value = variance_inflation_factor(dfn,column_index)
    print(column_index,"---------",vif_value)

    if vif_value <= 6:
        column_to_be_kept.append(numeric_columns[i])
        column_index = column_index +1 

    else:
        dfn = dfn.drop([numeric_columns[i]],axis=1)
```

```python
# First we need to create group of Resultant categorical column first
```

```python
# g1 = df[df["Approved_Flag"].str.contains("P1")]["Total_TL"]
# g1 = df[df["Approved_Flag"].str.contains("P2")]["Total_TL"]
# g1 = df[df["Approved_Flag"].str.contains("P3")]["Total_TL"]
# g1 = df[df["Approved_Flag"].str.contains("P4")]["Total_TL"]
```

```python

from scipy.stats import f_oneway
```

```python

kept_column_numeric = []

for i in column_to_be_kept:
    a = list(df[i])
    b = list(df["Approved_Flag"])

    g1 = [value for value, group in zip(a,b) if group =="P1"]
    g2 = [value for value, group in zip(a,b) if group =="P2"]
    g3 = [value for value, group in zip(a,b) if group =="P3"]
    g4 = [value for value, group in zip(a,b) if group =="P4"]

    f_statics,p_value = f_oneway(g1,g2,g3,g4)

    if p_value <=.05:
        kept_column_numeric.append(i)
```

```python
df[kept_column_numeric]
```

```python
# Another way to apply anova test

from scipy.stats import f_oneway

kept = []

for i in column_to_be_kept:

    g1 = df[df["Approved_Flag"] == "P1"][i]
    g2 = df[df["Approved_Flag"] == "P2"][i]
    g3 = df[df["Approved_Flag"] == "P3"][i]
    g4 = df[df["Approved_Flag"] == "P4"][i]

    f_stat, p_value = f_oneway(g1, g2, g3, g4)

    if p_value <= 0.05:
        kept.append(i)

print(kept)
```
